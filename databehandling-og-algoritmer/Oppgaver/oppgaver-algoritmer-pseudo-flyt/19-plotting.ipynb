```python
import json
import matplotlib.pyplot as plt

with open("land.json", encoding="utf-8") as fil:
    data = json.load(fil)

data[0] #sjekker at alt gikk fint med innlesing av fila

verdensdeler = {}
for land in data:
    if land['verdensdel'] not in verdensdeler:
        verdensdeler[land['verdensdel']] = 1
    else:
        verdensdeler[land['verdensdel']] += 1

print(verdensdeler)


```

```python
land = list(verdensdeler.keys()) # gjør nøklene til en liste
antall = list(verdensdeler.values()) # gjør verdiene til en lsite

# Oppgave: Plott land og antall
plt.bar(land, antall)
plt.title("Antall land på topplista")
plt.xlabel("Verdensdeler")
plt.ylabel("Antall")
plt.style.use("fivethirtyeight")
```
